original `AND gate`

```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1 * w1 + x2 * w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

```python
print(AND(0, 0))
print(AND(1, 0))
print(AND(0, 1))
print(AND(1, 1))
```

change theta to `-b`

`b`是偏置 != `w`是權重weight

偏置是調整神經元被激活的容易程度（輸出信號為1的程度），當`b` = -0.7，就代表信號的加權總和要超過0.7，神經元才會被激活

權重是控制輸入信號的重要性的程度

```python
import numpy as np

x = np.array([0, 1])
w = np.array([0.5, 0.5])
b = -0.7
print(w * x)
print(np.sum(w * x))
print(np.sum(w * x) + b)
```

```python
def AND(x1: float, x2: float) -> int:
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    return 0 if tmp <= 0 else 1
```

```python
def NAND(x1: float, x2: float) -> int:
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w*x) + b
    return 0 if tmp <= 0 else 1
```

```python
def OR(x1: float, x2: float) -> int:
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = 0.7
    tmp = np.sum(w*x) + b
    return 0 if tmp <= 0 else 1
```

單層感知器無法表示`XOR gate`/無法分離非線形空間，

但只要疊加不同邏輯控制的層就可以了

所以`XOR gate`就是2層感知器，而上面那些就是單層感知器

2層以上的感知器就叫多層感知器（multi-layered perceptron）

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```
